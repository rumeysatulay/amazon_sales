```python
# Importing necessary libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import matplotlib.dates as mdates
import ast
import math
import seaborn as sns

# Defining the path to the CSV files
file_path = r"C:\Users\rumey\OneDrive\Masaüstü\Projects\Amazon Sales\amazon.csv"


# Reading the CSV files
df = pd.read_csv(file_path, delimiter=',')
df
```

```python
# We can see the detailed information of the DataFrame. As shown, for 1,465 rows, there are no null values in any column.
# However, the values of discounted_price, actual_price, discount_percentage, rating, and rating_count are of type str.

df.info()
```

```python
df['discounted_price'] = df['discounted_price'].replace('[₹,]', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True).astype(float)
df['discount_percentage'] = df['discount_percentage'].replace('[%]', '', regex=True).astype(float)

#df['rating'] = df['rating'].astype(float) -- The transformation of the column types was not entirely successful because there are some values that cannot be converted.

df['rating_count'] = df['rating_count'].replace('[,]', '', regex=True).astype(float)
df
```

```python
# To make the analysis more understandable, the currency unit has been changed from rupees to dollars.

df['discounted_price'] = df['discounted_price']  / 83.77
df['actual_price'] = df['actual_price']  / 83.77
```

```python
# The presence of the '|' character in the rating column is preventing the data from being transformed.

df['rating'].unique()
```

```python
#To identify the row(s) where the rating is set to '|'.

df[df['rating'] == '|']
```

```python
# To delete that row: 

df = df.drop(1279)
```

```python
#Changing the type of the values string to numeric:

df['rating'] = df['rating'].astype(float)
df.info()
```

```python
df['rating'].value_counts().to_frame()
```

```python
#I created five rating categories: Very Good, Good, Okay, Bad, and Very Bad. 
# Additionally, I added a new column called evaluation to categorize the ratings.

def categorize_rating(rating):
    if rating >= 4.5:
        return 'Very Good'
    elif rating >= 4.0:
        return 'Good'
    elif rating >= 3.0:
        return 'Okay'
    elif rating >= 2.0:
        return 'Bad'
    else:
        return 'Very Bad'

# Apply the function to the 'rating' column
df['evaluation'] = df['rating'].apply(categorize_rating)
```

```python
# There are two null values in the rating_count column.

df.isnull().sum()
```

```python
# To ensure the accuracy of the analysis, we discard the rows that contain null values.

df = df.dropna(subset='rating_count')
```

```python
# If there are duplicate rows, I remove them.

df = df.drop_duplicates()
df
```

```python
# To see category column:

df['category']
```

```python
# The values in the category column have been consolidated from general to specific. 
# Let’s separate the items in this column using the & symbol.

df_split = df['category'].str.split('|', expand=True)
df_split.isnull().sum()
```

```python
df_split
```

```python
# The first two columns have no null values; therefore, I focus on these two columns as `cat_1` and `cat_2`.

df_split = df_split.drop(columns=[2,3,4,5,6])
df_split
```

```python
df_split.columns = ['cat_1', 'cat_2']
df_split
```

```python
# To merge the category DataFrames with the original DataFrame:

df = pd.concat([df, df_split], axis=1)
df
```

```python
pie_chart = df.groupby('cat_1')['cat_1'].agg('count').to_frame()
pie_chart.columns = ['count']
pie_chart = pie_chart.sort_values(by='count', ascending=False)
pie_chart.reset_index(inplace=True)
pie_chart.rename(columns={'cat_1': 'cat'}, inplace=True)
pie_chart
```

## Pie Chart of Products

```python
# Data to plot
labels = pie_chart['cat']
sizes = pie_chart['count']
colors = sns.color_palette('pastel')[0:len(labels)]

# Creating a pie chart
fig, ax = plt.subplots(figsize=(10, 8))
wedges, texts, autotexts = ax.pie(
    sizes, 
    colors=colors, 
    autopct='%1.1f%%', 
    startangle=140,
    pctdistance=0.85,
    labeldistance=1.2  # Adjusting the label distance
)

# Rotating the labels manually by 45 degrees
for text in texts:
    text.set_rotation(45)
    text.set_fontsize(14)
    text.set_color('black')

for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('white')
    

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  

# Adding a legend outside the pie
ax.legend(wedges, labels, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Categories of Products Distribution')
plt.show()
```

```python
cat1_count = df.groupby('cat_1')['rating_count'].agg('sum').to_frame().sort_values(by='rating_count', ascending=False)
cat2_count = df.groupby('cat_2')['rating_count'].agg('sum').to_frame().sort_values(by='rating_count', ascending=False)
cat1_count
```

## # of Reviews for Products according to Categories

```python
# Setting the style of seaborn plot
sns.set(style="whitegrid")

# Creating a figure and a set of subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 10))

# Ploting each barh chart in a subplot
sns.barplot(y='cat_1', x='rating_count', data=cat1_count, ax=axes[0], orient='h', palette='viridis').set(title='General Category', xlabel='# of Review of Products', ylabel='')
sns.barplot(y='cat_2', x='rating_count', data=cat2_count, ax=axes[1], orient='h', palette='plasma').set(title='Sub-category', xlabel='# of Review of Products', ylabel='')
ticks_x = plt.FuncFormatter(lambda y, pos: f'{int(y/1000000)}M')
axes[0].xaxis.set_major_formatter(ticks_x)
axes[1].xaxis.set_major_formatter(ticks_x)
# Adjusting the layout
plt.tight_layout()

plt.show()
```

```python
# Creating DataFrames by Grouping General and Subcategories to Show Price


cat_price = df[['product_id','actual_price','discounted_price','discount_percentage','cat_1','cat_2']]
cat1_price= cat_price.groupby('cat_1')[['actual_price','discounted_price','discount_percentage']].agg('mean').sort_values(by='actual_price', ascending=False)
cat2_price= cat_price.groupby('cat_2')[['actual_price','discounted_price','discount_percentage']].agg('mean').sort_values(by='actual_price', ascending=False)
cat2_price
```

## Mean Price of Products according to Categories

```python
sns.set(style="whitegrid")

# Creating a figure and a set of subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 10))
sns.barplot(y='cat_1', x='actual_price', data=cat1_price, ax=axes[0], orient='h', palette='inferno').set(title='Mean Price of Categories', xlabel='Price', ylabel='')
sns.barplot(y='cat_2', x='actual_price', data=cat2_price, ax=axes[1], orient='h', palette='inferno').set(title='Mean Price of Sub-Categories', xlabel='Price', ylabel='')
ticks_x = plt.FuncFormatter(lambda y, pos: f'${int(y)}')
axes[0].xaxis.set_major_formatter(ticks_x)
axes[1].xaxis.set_major_formatter(ticks_x)

plt.tight_layout()
plt.show()
```

## Rating Evaluation

```python
# Plotting
plt.figure(figsize=(10, 6))
custom_palette = ['#003366', '#004080', '#00509e', '#0066cc', '#3399ff']
bar_chart = sns.barplot(data=df_sorted, x='evaluation', y='count', palette=custom_palette)

# Customizing the plot
bar_chart.set_title('Rating & Rating Counts')
bar_chart.set_xlabel('Evaluation')
bar_chart.set_ylabel('Rating Counts')

for p in bar_chart.patches:
    height = p.get_height()
    bar_chart.annotate(format(height, '.0f'),
                       (p.get_x() + p.get_width() / 2., height),
                       ha='center', va='center',
                       xytext=(0, 5),  # 5 points vertical offset
                       textcoords='offset points')

# Rotating x-axis labels if needed
bar_chart.set_xticklabels(bar_chart.get_xticklabels(), rotation=45, ha='right')

plt.show()
```
